```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
x= newdf.drop('stroke',axis=1)
# axis =1 means column dropped and axis=0 means row dropped
y= newdf['stroke']
#now splitting of data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
# splitting the data into two modules train and testby random_state=100(randomly tasking 100 data) test_size=0.2 meaning training data 80 and testing data 20
logmodel=LogisticRegression()
#model calling and storing into variable
logmodel.fit(x_train,y_train)
#prediction with comparison to x_test values(80:20)
predictions=logmodel.predict(x_test)
#printing all data
print('------------------------------predicting-------------------------------\n')
print(confusion_matrix(y_test,predictions))
print('------------------------------scores-----------------------------------\n')
print("accuracy is::",accuracy_score(y_test,predictions)*100)
#accuracy is total correct predictions / total predictions mode * 100
print("precision is::",precision_score(y_test,predictions)*100)
print("recall is::",recall_score(y_test,predictions)*100)
#recall is true postiive/ true positive + false negative
#true positive of occuring correct events
#false negative of occuring false events
print("f1_score is::",f1_score(y_test,predictions)*100)
#f1_score conveys the balance between precision and recall, f1= (2*precision *recall / (precision + recall))
print("y_test values number of 0's and 1's for stroke");
print(y_test.value_counts())
predictions=pd.Series(predictions)
print('----------------------------predicting-----------------------------\n')
print("Prediction Values for stokes:-")
print(predictions.value_counts())
```

# KNN REGRESSION MODEL for training and predictions


```python
from sklearn.neighbors import KNeighborsClassifier
x= newdf.drop('stroke',axis=1)
# axis =1 means column dropped and axis=0 means row dropped
y= newdf['stroke']
#now splitting of data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
#lazy learner algorithm
K_model = KNeighborsClassifier(n_neighbors=99,p=2,metric='euclidean')
K_model.fit(x_train,y_train)
predictions = K_model.predict(x_test)
#printing all data
print('---------------KNN model prediction--------------------')
print('------------------------------predicting-------------------------------\n')
print(confusion_matrix(y_test,predictions))
print('------------------------------scores-----------------------------------\n')
print("accuracy is::",accuracy_score(y_test,predictions)*100)
#accuracy is total correct predictions / total predictions mode * 100
print("precision is::",precision_score(y_test,predictions)*100)
print("recall is::",recall_score(y_test,predictions)*100)
#recall is true postiive/ true positive + false negative
#true positive of occuring correct events
#false negative of occuring false events
print("f1_score is::",f1_score(y_test,predictions)*100)
#f1_score conveys the balance between precision and recall, f1= (2*precision *recall / (precision + recall))
print("y_test values number of 0's and 1's for stroke");
print(y_test.value_counts())
predictions=pd.Series(predictions)
print('----------------------------predicting-----------------------------\n')
print("Prediction Values for stokes:-")
print(predictions.value_counts())
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn import model_selection
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import model_selection
x=df.drop(['stroke','Residence_type'],axis=1)
y=df['stroke']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
seed=100
kfold=model_selection.KFold(n_splits=10,random_state=None)
estimators=[]
logrec=LogisticRegression()
logrec.fit(x_train,y_train)
estimators.append(('logistic',logrec))
dectree=DecisionTreeClassifier()
dectree.fit(x_train,y_train)
estimators.append(('decisiontree',dectree))
kneigh=KNeighborsClassifier(n_neighbors=35,p=2,metric='euclidean')
kneigh.fit(x_train,y_train)
estimators.append(('knearestneighbour',kneigh))
ganb=GaussianNB()
ganb.fit(x_train, y_train)
estimators.append(('naivebayesgauss',ganb))
svm_model = SVC(kernel = 'linear', random_state=0)
svm_model.fit(x_train,y_train)
estimators.append(('SVM',svm_model))
ensemble=VotingClassifier(estimators)
results=model_selection.cross_val_score(ensemble,x_test,y_test,cv=kfold)
print('Voting result')
print('------------------result generating---------------------')
print(results)
print('\n')
print(results.mean()*100)
print('\n')
model = BaggingClassifier()
# evaluate the model
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model, x, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Bagging result')
print('------------------result generating---------------------')
print(n_scores)
print('\n')
print(n_scores.mean()*100)
```
